```julia
for i=1:10
    data=npzread("tilepos" * string(i) * ".npy")
    mc = cumulants(data, 4)
    inp2 = Array(mc[4])
    out = cp_als(inp2,1)
    f(w) = sum(w.^4)/(sum(w.^2).^2)
    w=out.fmat[1]
    println(f(w))
end
```
